```python
import matplotlib.pyplot as plt
import numpy as np
import subprocess
import os
```

```python
os.system("rm *.png")
result = subprocess.run(['ls'], stdout=subprocess.PIPE)
scans=result.stdout.decode('utf-8')
scans
```

```python
scans=scans.split('\n')[:-1]
scans.remove('plotter.ipynb')
```

```python
for scan in scans:
    print(scan)
    # read from csv
    data = np.genfromtxt(scan, delimiter=',')
    # plot
    result = subprocess.run(['head','-20',f'{scan}'], stdout=subprocess.PIPE)
    head=result.stdout.decode('utf-8')
    head=head.split('\n')[:-1]
    plt.title(scan[:-4])
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Intensity (a.u.)')
    for i in range(1,len(head)):
        plt.text(0.5, i*0.05, head[i][:-2], transform=plt.gca().transAxes,fontsize=8, color='gray')
        # show head as text
    plt.plot(data[:,0], data[:,1])
    plt.show()
    #save
    plt.savefig(f'{scan[:-4]}.png', transparent=False)

```
